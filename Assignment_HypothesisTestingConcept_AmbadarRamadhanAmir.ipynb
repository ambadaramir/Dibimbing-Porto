# 2. T-test

Dua data diatas adalah data tekanan darah (blood pressure) dari pasien sebelum dan sesudah treatment. <br>
gunakan alpha = 5% untuk menentukan apakah tekanan darah setelah treatment menjadi lebih tinggi dibanding sebelum treatment.

```python
### COMPLETE THE FOLLOWING CODE (score : 40)
## select doubt level of alpha
alpha = 0.05

## what kind of t-test is it? (two sample, paired-two sample or one sample)
# Answer : paired-two sample

## select alternative hypothesis ('two-sided', 'greater' or 'less')
alt = 'greater'

## calculate using packages scipy.stats
stat, pval = stats.ttest_rel(list_bp_after, list_bp_before, alternative=alt)

## give conclusion using if else condition
print('p-value =', pval)
if pval>alpha/2:
  print("Conclusion : pasien setelah treatment tidak mempengaruhi tekanan darah")
else:
  print("Conclusion : pasien setelah treatment mempengaruhi tekanan darah")
```
